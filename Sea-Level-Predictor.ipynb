```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
def draw_plot():
    # Read data from file 
    df = pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    fig, ax = plt.subplots()
    
    ax.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'])

    
    # Create first line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])

    plt.plot(df['Year'], slope * df['Year'] + intercept, color='red', label=f'Best fit line: y = {slope:.2f}x + {intercept:.2f}')
   
    # Create second line of best fit
    years_2000 = df[df['Year'] >= 2000]['Year']
    sea_levels_2000 = df[df['Year'] >= 2000]['CSIRO Adjusted Sea Level']


    slope_2000, intercept_2000, r_value, p_value, std_err = linregress(years_2000, sea_levels_2000)

    plt.plot(years_2000, slope_2000 * years_2000 + intercept_2000, color='green', label=f'Best fit line from 2000: y = {slope_2000:.2f}x + {intercept_2000:.2f}')

    
    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()

    plt.grid(True)
    plt.show()
```

```python
draw_plot()
```
